```python
import requests
import numpy as np
import json
```

```python
payload = {
    "inputs": [
        {
            "name": "TEXT",
            "shape": [1],
            "datatype": "BYTES",
            "data": ["Đây là văn bản đầu vào của tôi"]
        }
    ]
}
```

```python
url = "http://localhost:8000/v2/models/viencoder/infer"
response = requests.post(url, json=payload)

if response.status_code == 200:
    print(response.json())
else:
    print("Inference failed:", response.status_code, response.text)
```
